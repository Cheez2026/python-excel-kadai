```python
import pandas as pd
import openpyxl
from openpyxl.styles import PatternFill

# Excelファイルを読み込む
df_2022 = pd.read_excel("2022_年間売上表.xlsx")
df_2023 = pd.read_excel("2023_年間売上表.xlsx")

# データを連結
df = pd.concat([df_2022, df_2023])

# 商品・売上年ごとにグループ化して合計を算出
df_grouped = df.groupby(["商品", "売上年"])["金額（千円）"].sum().reset_index()

# Excelに書き出し
df_grouped.to_excel("売上集計表.xlsx", index=False)

# ヘッダーを薄いグレーに設定
wb = openpyxl.load_workbook("売上集計表.xlsx")
ws = wb.active

fill = PatternFill(fill_type="solid", fgColor="F2F2F2")
for cell in ws[1]:
    cell.fill = fill

wb.save("売上集計表.xlsx")
print("売上集計表.xlsx を作成しました")
```
