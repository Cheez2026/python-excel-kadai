```python
import openpyxl
from datetime import datetime

# ワークブック作成
wb = openpyxl.Workbook()
ws = wb.active

# 現在日付
today = datetime.now().strftime("%Y/%m/%d")
filename = f"請求書_{datetime.now().strftime('%Y%m%d')}.xlsx"

# 請求書タイトル
ws["B2"] = "請求書"

# 会社情報
ws["B4"] = "株式会社ABC"
ws["B5"] = "〒101-0022 東京都千代田区神田練塀町300"
ws["B6"] = "TEL:03-1234-5678 FAX:03-1234-5678"
ws["B7"] = "担当者名:鈴木一郎 様"

# No.と日付
ws["F4"] = "No."
ws["G4"] = "0001"
ws["F5"] = "日付"
ws["G5"] = today

# 商品テーブルのヘッダー
ws["B10"] = "商品名"
ws["C10"] = "数量"
ws["D10"] = "単価"
ws["E10"] = "金額"

# 商品データ
items = [
    ("商品A", 2, 10000),
    ("商品B", 1, 15000),
]

for i, (name, qty, price) in enumerate(items):
    row = 11 + i
    ws[f"B{row}"] = name
    ws[f"C{row}"] = qty
    ws[f"D{row}"] = price
    ws[f"E{row}"] = qty * price

# 合計行
ws["E13"] = 35000

# 小計・消費税・合計
ws["B15"] = "小計"
ws["E15"] = 35000
ws["B16"] = "消費税"
ws["E16"] = 3500
ws["B17"] = "合計"
ws["E17"] = 38500

# 保存
wb.save(filename)
print(f"{filename} を作成しました")
```
